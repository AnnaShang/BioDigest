```python
y2 = hainan['Raw Biogas Produced (m3)']
mod = sm.tsa.statespace.SARIMAX(y2, \
                                order=(1, 1, 1), seasonal_order=(1, 1, 1, 12),\
                                enforce_stationarity=False, enforce_invertibility=False)

results = mod.fit()

print(results.summary().tables[1])
```
